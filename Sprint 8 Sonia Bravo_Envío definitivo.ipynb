```python
from sqlalchemy import text
from sqlalchemy import create_engine
import pandas as pd

# Crear el engine
engine = create_engine("mysql+pymysql://root:rootroot@localhost/transactions_3")

# Obtener nombres de todas las tablas
with engine.connect() as conn:
    tablas = conn.execute(text("SHOW TABLES")).fetchall()


# Crear diccionario con un DataFrame por tabla
Trans3 = {}
for tabla in tablas:
    nombre = tabla[0]
    df = pd.read_sql(f"SELECT * FROM {nombre}", con=engine)
    Trans3[nombre] = df

# Ejemplo: mostrar las primeras filas de una tabla
print(Trans3['transactions'].head())
```

- Exercici 6 Tres variables.

```python
Trans3['transactions']['Año'] = Trans3['transactions']['timestamp'].dt.year

Tot= pd.merge(Trans3['transactions'],Trans3['users'], left_on='user_id', right_on='id') 

grouped = Tot.groupby(['country', 'Año'])['amount'].sum().reset_index()

Colores= { 2021: 'darkblue',2022: 'lightblue'}

grafico=sns.barplot(data=grouped, x='country', y='amount', hue='Año', palette= Colores, alpha=0.7)

for container in grafico.containers:
    grafico.bar_label(container, labels=[f'{int(bar.get_height()):,}'.replace(',', '.') for bar in container],label_type='edge')
# Punto de miles en el eje y

import matplotlib.ticker as mticker
grafico.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'.replace(',', '.')))

plt.title('Total de ventas por países y año', fontsize=13)
plt.xlabel('Países', fontsize=11)
plt.ylabel('Amount', fontsize=11)
plt.xticks(rotation=360)
plt.show()
```
